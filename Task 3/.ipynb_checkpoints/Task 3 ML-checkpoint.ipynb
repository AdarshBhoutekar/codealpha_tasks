```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load MNIST CSVs
mnist_train = pd.read_csv("mnist_train.csv")
mnist_test  = pd.read_csv("mnist_test.csv")

print("Train shape:", mnist_train.shape)
print("Test shape :", mnist_test.shape)
```

```python
X_train = mnist_train.iloc[:, 1:].values
y_train = mnist_train.iloc[:, 0].values

X_test = mnist_test.iloc[:, 1:].values
y_test = mnist_test.iloc[:, 0].values
```

```python
X_train = X_train / 255.0
X_test  = X_test / 255.0
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test  = X_test.reshape(-1, 28, 28, 1)

print(X_train.shape)
print(X_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat  = to_categorical(y_test, num_classes=10)
```

```python
plt.imshow(X_train[0].reshape(28,28), cmap="gray")
plt.title(f"Label: {y_train[0]}")
plt.axis("off")
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Dense, Flatten,
    Dropout, BatchNormalization
)
```

```python
model = Sequential([

    # Block 1
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),

    # Block 2
    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),

    # Fully connected layers
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(10, activation='softmax')  # 10 digits
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

lr_reduce = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=2
)
```

```python
history = model.fit(
    X_train, y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=15,
    batch_size=128,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"Test Accuracy: {test_acc * 100} %")
```
